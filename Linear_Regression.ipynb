üî∏ Step 1: üìò Concept Summary

    Linear Regression ek supervised learning algorithm hai jisme hum input features ke basis pe output ko predict karte hain ‚Äî especially jab output continuous hota hai (jaise price, salary, temperature).

Supervised Learning kyun?

    Kyunki training data ke saath output (label) already hota hai, jisko model seekh sakta hai.

Regression kyun?

    Kyunki hum numeric (continuous) value predict kar rahe hain ‚Äî classification nahi

üî∏ Step 2: üî£ Mathematical Intuition

        y = mx + c

        y: predicted value

        m: slope (kitna fast y change ho raha x ke respect me)

        x: input

        c: intercept (line kaha se start ho rahi hai)

        Ham chahte hain ki hamari line data ke bilkul pass se guzre ‚Äî jise bolte hain line of best fit.

        Iske liye cost function minimize karte hain ‚Äî Mean Squared Error (MSE).

üîç Kya hota hai Gradient Descent?
        Ek iterative optimization technique hoti hai jo model ke weights (m, c) ko adjust karti hai taki cost function (error) ko minimum kiya ja sake.

üìà Kaise kaam karta hai?
        Maan lo cost function ek hill hai.
        Gradient descent slowly downhill chalta hai, jahan cost sabse kam ho ‚Äî wahan rukta hai.
Step size ko bolte hain: Learning Rate.

Gradient Descent matlab "error ko dheere dheere kam karte jao jab tak model best na ban jaaye".

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# üìä Dummy Data Banana
# Simple dataset
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # input features
y = np.array([2, 4, 6, 8, 10])               # target output

```

```python
# ‚úÇÔ∏è Step 3: Train-Test Split


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)

#Yahan hum kya kar rahe hain? Data ko 80-20 ratio me split kiya training aur testing ke liye random_state=42 repeatability ke liye fix kiya.
```

```python
#üèóÔ∏è Step 4: Model Training
model = LinearRegression()
model.fit(X_train, y_train)

#Model train ho gaya ‚Äî ab yeh X aur y ke beech relation samajh gaya hai.
```

```python
# Step 5: Predictions and Evaluation
y_pred = model.predict(X_test)
print("Predicted:", y_pred)
print("MSE:", mean_squared_error(y_test, y_pred))
print("R2 Score:", r2_score(y_test, y_pred))

'''Model ko test data diya, usne output predict kiya.
mean_squared_error: jitni prediction aur actual me difference hai uska square.
r2_score: kitna acha model fit hua ‚Äî 1.0 perfect, 0.0 worst.
R2 Score jitna 1 ke kareeb, utna model strong. MSE jitna kam, utna better.'''
```

```python
#üìà Step 6: Visualization

plt.scatter(X, y, color='blue')
plt.plot(X, model.predict(X), color='red')
plt.title("Linear Regression Line")
plt.xlabel("X")
plt.ylabel("y")
plt.grid(True)
plt.show()

'''Blue dots: actual data 
Red line: model ka prediction (line of best fit) '''
```
